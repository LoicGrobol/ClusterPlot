# Imports

```python
import sys

sys.path.append("../ClusterPlot")
sys.path.append("./utils")

from DataSetFactory import DataSetFactory
from cluster_plot import ClusterPlot
```

```python
RANDOM_STATE = 42
```

# Original Dim

```python
ds = DataSetFactory.get_dataset("MNIST_AE_ORIG", RANDOM_STATE)
```

```python
X = ds.df[ds.feature_cols].values
y = ds.df[ds.label_col].values
```

```python
orig_dim_cp = ClusterPlot(
    learning_rate=0.5,
    n_iter=41,
    batch_size=1,
    anchors_method="birch",
    birch_threshold=5.6,
    dataset="MNIST",
    class_to_label=ds.class_to_label,
    show_fig=True,
    save_fig=False,
    k=200,
    random_state=RANDOM_STATE,
    save_fig_every=40,
    magnitude_step=True,
    top_greedy=1,
    alpha=0.3,
    remove_outliers_k=10,
    douglas_peucker_tolerance=0.3,
    smooth_iter=3,
)
print(orig_dim_cp)
```

```python
low_dim_orig_dim = orig_dim_cp.fit_transform(X, y)
```
